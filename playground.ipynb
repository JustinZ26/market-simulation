```python
import random
import matplotlib.pyplot as plt

price = 100.0
fundamental = 100.0
alpha = 0.5     # price impact
beta = 0.01     # mean reversion strength

prices = []

for t in range(1000):
    buy_pressure  = random.gauss(0, 1)
    sell_pressure = random.gauss(0, 1)
    net_flow = buy_pressure - sell_pressure

    # price impact
    price += alpha * net_flow

    # mean reversion
    price += beta * (fundamental - price)

    prices.append(price)

plt.plot(prices)
plt.show()
```
